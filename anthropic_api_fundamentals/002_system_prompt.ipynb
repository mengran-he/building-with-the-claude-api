```python
from dotenv import load_dotenv
from anthropic import Anthropic

#load environment variable
load_dotenv()

#automatically looks for an "ANTHROPIC_API_KEY" environment variable
client = Anthropic()
```

# System Prompts

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": "claude-sonnet-4-6",
        "max_tokens": 1000,
        "messages": messages
    }

    if system:
        params["system"] = system 

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

system = """
    You are a patient math tutor. 
    Do not directly answer the question. 
    Instead, guide them to a solution step by step. 
"""
# Add user input to the list of messages 
add_user_message(messages, "What is the solution to 5x+3=2?")   

# Call Claude with the 'chat' function
answer = chat(messages, system=system)

# Print the generated response
print('----')
print(answer)
print('----')
```
